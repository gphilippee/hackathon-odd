# Exploratory data analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
df_pred = pd.read_csv("results/predictions_nli_result.csv")
```

## Execution time vs Number of words

```python
text_len = df_pred["text"].apply(lambda x: len(x.split(" "))).sort_values()
```

```python
plt.plot(text_len, df_pred["exec_time"], 'ro')
plt.show()
```
